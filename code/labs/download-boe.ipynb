```python
import os
import time

def download_boe(n=0):
    anio = 1960

    # Create the directory if it doesn't exist
    file_path = f"../data/BOE/{anio}/"
    if not os.path.exists(file_path):
        os.makedirs(file_path)

    for i in range(n, 999999):
        url = f"https://www.boe.es/diario_boe/xml.php?id=BOE-A-{anio}-{i}"
        print(f"[>][URL] {url}")
        
        try:
            response = requests.get(url, timeout=10)
        except Exception as e:
            return i

        #error?
        if '<error>' in response.text:
            print(f"[X][ERROR] {url}")
            break
        # Define the file path
        file_path = f"../data/BOE/{anio}/BOE-A-{anio}-{i}"    
        # Save the XML response to the file
        with open(file_path, 'w') as file:
            file.write(response.text)

        time.sleep(0.5)

    return -1

i = download_boe(458)
while i != -1:
    print(f"[>][CONTINUE] {i}")
    time.sleep(30)
    i = download_boe(i)

```
